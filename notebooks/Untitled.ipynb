```python
pldf = pl.read_csv(paths[0], try_parse_dates=True)
```

```python
shcode = pldf.group_by('tr_key').agg().to_dicts()
```

```python
for code in shcode:
    a = pldf.filter(pldf['tr_key'] == code['tr_key'])
```

```python
a.rename({'tr_cd': 'cd'})
```
